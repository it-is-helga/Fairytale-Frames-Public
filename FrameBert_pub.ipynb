### Imports & Frame Predictor model

```python
frame_model = 'liyucheng/frame_finder'
```

```python
import numpy as np
import pandas as pd
import nltk
nltk.download('punkt_tab')
```

```python
#FRAME PREDICTOR

import torch
from transformers import RobertaForTokenClassification, RobertaTokenizerFast
from nltk.tokenize import sent_tokenize, word_tokenize

# Load pre-trained model and tokenizer
model = RobertaForTokenClassification.from_pretrained(frame_model)
tokenizer = RobertaTokenizerFast.from_pretrained('roberta-base', add_prefix_space=True)

device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
model.to(device)
model.eval()

def predict_frames(input_text):
    """
    Predict frames for the input text.

    Args:
    - input_text (str): The text for which to predict frames.

    Returns:
    - List of frames predicted for each token in the input text.
    """
    # Tokenize input text at sentence level
    input_text = str(input_text)
    sentences = sent_tokenize(input_text)
    predictions = []

    for sentence in sentences:
        # Tokenize and prepare input for the model
        tokenized = tokenizer(sentence, truncation=True, return_tensors="pt").to(device)

        # Model inference
        with torch.no_grad():
            outputs = model(**tokenized)

        logits = outputs.logits
        predictions_ids = torch.argmax(logits, dim=2).squeeze().cpu().numpy()

        # Decode predictions
        tokens = tokenizer.convert_ids_to_tokens(tokenized['input_ids'].squeeze().cpu().numpy())
        # print(tokens)
        sentence_predictions = [(token, frame_list[pred]) for token, pred in zip(tokens, predictions_ids)]
        predictions.append(sentence_predictions)
        s = set()
        for sentence_prediction in predictions:
            # print(sentence_prediction)
            s1 = {t[1] for t in sentence_prediction}
            s = set.union(s, s1)

    return s
```

### Data retrieval

```python
url = "https://github.com/it-is-helga/Fairytale-Frames-Public/raw/refs/heads/main/data/Santal_Demo.xlsx"
df = pd.read_excel(url)
df.head()
```

### Frame extraction

```python
# apply frames to dataset
target_column = 'Text' # specify text column title
df['frames'] = df[target_column].apply(predict_frames)
```

```python
# OHE
# Create a list of all unique elements across sets
all_categories = set.union(*df['frames'])

# Create one-hot encoded columns
new_columns = {
    category: df['frames'].apply(lambda x: 1 if category in x else 0)
    for category in all_categories
}
df = pd.concat([df, pd.DataFrame(new_columns)], axis=1)
## Drop the original set column if no longer needed
# df = df.drop(columns=['Categories'])
```
